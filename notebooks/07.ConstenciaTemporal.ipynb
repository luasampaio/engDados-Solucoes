```python
import pandas as pd

# converte para datetime (UTC) e lida com valores inválidos
df["evendata_cadastro"] = pd.to_datetime(df["data_cadastro"], utc=True, errors="coerce")

agora = pd.Timestamp.now("UTC")
df["anomalia"] = (
    df["event_time"].isna() |
    (df["event_time"] < agora - pd.Timedelta(days=365)) |
    (df["event_time"] > agora + pd.Timedelta(days=1))
)

print(df.loc[df["anomaly"], ["timestamp", "event_time"]])
```
