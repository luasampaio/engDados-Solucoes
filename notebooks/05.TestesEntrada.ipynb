```python
def extrair_usuario_email_provedor(email: str) -> (): # type: ignore
    email_separado = email.split(sep='@')
    usuario = email_separado[0]
    provedor = email_separado[1]
    return usuario, provedor
```

```python
df.columns
```

```python
# cria as duas colunas a partir da coluna "email"
df[["usuario", "provedor"]] = (
    df["email"].astype(str)
        .apply(extrair_usuario_email_provedor)  # -> tupla (usuario, provedor)
        .apply(pd.Series)                       # expande para 2 colunas
)
```
