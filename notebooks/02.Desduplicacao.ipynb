```python
import pandas as pd
import numpy as np

def relatorio_nulos(df: pd.DataFrame, considerar_textos_vazios: bool = False) -> pd.DataFrame:
    """
    Devolve um DataFrame com a porcentagem de nulos por coluna (ordenado).
    Se considerar_textos_vazios=True, "" e espaços em branco contam como nulo.
    """
    dados = df.copy()

    if considerar_textos_vazios:
        obj = dados.select_dtypes(include=["object", "string"]).columns
        if len(obj):
            # transforma strings vazias ou só com espaços em NA
            dados[obj] = dados[obj].astype("string").apply(lambda c: c.str.strip().replace({"": pd.NA}))

    pct = dados.isna().mean().sort_values(ascending=False)
    return pct.rename("pct_nulos").to_frame()

def colunas_suspeitas_por_nulos(df: pd.DataFrame, limite: float = 0.20, considerar_textos_vazios: bool = False) -> pd.Series:
    """
    Retorna apenas as colunas com porcentagem de nulos acima do limite.
    """
    rel = relatorio_nulos(df, considerar_textos_vazios=considerar_textos_vazios)
    return rel.query("pct_nulos > @limite")["pct_nulos"]
```

### Print para Validar o percentual de dados nulos

```python
rel = relatorio_nulos(df, considerar_textos_vazios=True)
print(rel.head(10))
suspeitas = colunas_suspeitas_por_nulos(df, limite=0.20, considerar_textos_vazios=True)
print("Colunas com >20% nulos:\n", suspeitas)
```
