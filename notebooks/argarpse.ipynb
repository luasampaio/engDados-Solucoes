```python
import argparse

def get_parser():
    p = argparse.ArgumentParser(
        prog="meu_app",
        description="Exemplo simples de argparse",
    )
    p.add_argument("entrada", help="arquivo de entrada")
    p.add_argument("-o", "--output", help="arquivo de sa√≠da", default="out.txt")
    p.add_argument("-v", "--verbose", action="count", default=0,
                   help="aumenta verbosidade (-v, -vv, ...)")
    p.add_argument("--limite", type=int, default=100, metavar="N")
    return p

if __name__ == "__main__":
    args = get_parser().parse_args()
    print(args.entrada, args.output, args.verbose, args.limite)
```
