### Teste velocidade Pyarrow ao ler dados

```python
csv_path = 'E:/engDados-Solucoes/datasets/staffs.csv'
```

```python
import time
import pandas as pd

def load_with_time(csv_path):
    """
    Lê um CSV com pandas e retorna (df, tempo_em_segundos).
    Também imprime o tempo formatado.
    """
    start_time = time.perf_counter()
    df = pd.read_csv(csv_path, low_memory=False)
    total_time = time.perf_counter() - start_time
    print(f"tempo total: {total_time:0.02f}s")
    return df, total_time
```

```python
import time
import pandas as pd

def load_with_time_fast(csv_path, **read_csv_kwargs):
    """
    Tenta ler com engine='pyarrow' (se disponível) e mede o tempo.
    Se algum argumento não for compatível com pyarrow, faz fallback para o engine padrão.
    Retorna (df, elapsed_seconds).
    """
    start = time.perf_counter()
    try:
        # Remover opções incompatíveis com pyarrow
        incompatible = {"low_memory", "memory_map", "float_precision"}
        clean_kwargs = {k: v for k, v in read_csv_kwargs.items() if k not in incompatible}
        df = pd.read_csv(csv_path, engine="pyarrow", **clean_kwargs)
    except Exception as e:
        # Fallback para engine padrão do pandas (C)
        df = pd.read_csv(csv_path, **read_csv_kwargs)
    elapsed = time.perf_counter() - start
    print(f"tempo total: {elapsed:0.02f}s")
    return df, elapsed
```

```python
df, t = load_with_time_fast(csv_path)
```

```python
df, t = load_with_time(csv_path)
```
